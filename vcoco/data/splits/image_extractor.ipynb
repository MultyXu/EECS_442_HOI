```python
# Step 1: Read IDs from Files
def move(set_name):
  train_lists = []

  # read all files
  base_dir = '/content/drive/MyDrive/EECS 442 final project/v-coco/data/splits'
  file_name = os.listdir(base_dir)[0]
  train_dir = os.path.join(base_dir, 'vcoco_{:s}.ids'.format(set_name))
  with open(train_dir, 'r') as file:
      ids = [int(line.strip()) for line in file]
      train_lists.extend(ids)

  # Step 2: Load COCO Dataset
  coco = COCO('/content/drive/MyDrive/EECS 442 final project/v-coco/data/instances_vcoco_all_2014.json'.format(set_name))  # Replace with the path to your COCO annotations file

  # Step 3: Sort Images Using IDs
  sorted_images = []
  for img in coco.dataset['images']:
    if img['id'] in train_lists:
      # print(img['id'], 'match')
      sorted_images.append(img)
  # print(len(sorted_images))
  # print(len(train_lists))
  # for img in coco.dataset['images']:
  #   print(img['id'])

  # for i in train_lists:
  #   print(i)

  # Step 4: Move or Copy Images to /image/train directory
  output_directory = '/content/drive/MyDrive/EECS 442 final project/coco_converted/images/{:s}'.format(set_name)
  train_path = '/content/drive/MyDrive/EECS 442 final project/trainval2014'
  # val_path = '/content/drive/MyDrive/EECS 442 final project/val2014'

  for img_info in sorted_images:
      img_filename = img_info['file_name']

      img_path = os.path.join(train_path, img_filename)
      output_path = os.path.join(output_directory, img_filename)
      if(os.path.exists(img_path)):
        shutil.move(img_path, output_path)

move("train")
```

```python
def move_images_by_set(set_name, image_ids, coco, source_dir, output_dir):
    set_images = [img for img in coco.dataset['images'] if img['id'] in image_ids]

    for img_info in set_images:
        img_filename = img_info['file_name']
        img_path = os.path.join(source_dir, img_filename)
        output_path = os.path.join(output_dir, set_name, img_filename)

        # Choose either shutil.copy or shutil.move based on your preference
        shutil.copy(img_path, output_path)
        # shutil.move(img_path, output_path)

# Step 1: Read JSON files for train, test, and val
train_json_path = '/content/drive/MyDrive/EECS 442 final project/v-coco/data/train.json'
# test_json_path = 'path/to/test.json'
# val_json_path = 'path/to/val.json'

with open(train_json_path, 'r') as file:
    train_data = json.load(file)

# with open(test_json_path, 'r') as file:
#     test_data = json.load(file)

# with open(val_json_path, 'r') as file:
#     val_data = json.load(file)

# Extract image IDs for each set
train_image_ids = [entry['id'] for entry in train_data['images']]
# test_image_ids = [entry['id'] for entry in test_data['images']]
# val_image_ids = [entry['id'] for entry in val_data['images']]

# Step 2: Load COCO Dataset
coco = COCO('/content/drive/MyDrive/EECS 442 final project/annotations/instances_train2017.json')  # Replace with the path to your COCO annotations file

# Step 3: Sort and Move/Copy Images for Train, Test, and Val
source_images_dir = '/content/drive/MyDrive/EECS 442 final project/train2017'  # Replace with the path to your COCO images directory
output_base_dir = '/content/drive/MyDrive/EECS 442 final project/coco_converted/images'   # Replace with the base path for your output directories

move_images_by_set('train', train_image_ids, coco, source_images_dir, output_base_dir)
# move_images_by_set('test', test_image_ids, coco, source_images_dir, output_base_dir)
# move_images_by_set('val', val_image_ids, coco, source_images_dir, output_base_dir)
```
